```python
df_raw = (spark.read
          .option("header", "true")
          .option("inferSchema", "true")
          .csv("/Volumes/workspace/default/loan_risk/loan.csv"))

print(f"Rows   : {df_raw.count():,}")
print(f"Columns: {len(df_raw.columns)}")
```

```python
spark.sql("CREATE DATABASE IF NOT EXISTS Loan_risk")
spark.sql("USE Loan_risk")
print("Database created")
```

```python
(
    df_raw.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("loan_risk.bronze_loans")
)

print("Bronze table saved!")
```

```python
bronze = spark.table("loan_risk.bronze_loans")
print(f"Row count : {bronze.count():,}")
print(f"Columns  : {len(bronze.columns)}")
display(bronze.limit(5))
```

```python
display(spark.sql("DESCRIBE HISTORY loan_risk.bronze_loans"))
```
